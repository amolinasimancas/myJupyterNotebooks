### Ejemplos de desplazamientos

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000

def f(x):
  return x**2

c = 5

x = np.linspace(-10,10, num=N)

y1 = f(x + c)
y2 = f(x - c)
y3 = f(x) + c
y4 = f(x) - c

fig, ax = plt.subplots()
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
ax.plot(x,y4)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

### Ejemplos de alargamientos y compresiones

```python
N = 1000

def f(x):
  return np.sin(x)

c = 2

x = np.linspace(-15,15, num=N)

y1 = f((1/c)*x)
y2 = f(c*x)
y3 = c*f(x)
y4 = (1/c)*f(x)


fig, ax = plt.subplots()
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
ax.plot(x,y4)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

### Ejemplos de reflexiones

```python
N = 1000

def f(x):
  return x**3

x = np.linspace(-10,10, num=N)

y1 = f(-x)
y2 = f(x)

fig, ax = plt.subplots()
ax.plot(x,y1)
ax.plot(x,y2)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```
